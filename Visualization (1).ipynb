```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('branch level revenue %.csv')
```

```python
df
```

```python


plt.figure(figsize=(10,5))
sns.barplot(x='Branch', y='branch_revenue', data=df, palette= 'bright',hue='Branch')
plt.xticks(rotation=45)
plt.xlabel('Branch')
plt.ylabel('Revenue')
plt.title('Branch Revenue')
plt.show()
```

```python

plt.figure(figsize=(10,5))
plt.pie(df['branch_revenue'], labels=df['Branch'], autopct='%1.1f%%')
plt.title('Branch Revenue Percentage')
plt.show()
```

#Insights:
Palayam is the primary revenue driver, contributing 52.8% of total revenue, while Muttathara adds 35.1% and Counter Sale contributes 12.1%. This highlights strong branch leadership at Palayam but also reveals revenue dependency risk, creating an opportunity to strengthen secondary branches for better balance and resilience.

```python

bf = bf[bf['year_month'] != '2026-01']
```

```python

plt.figure(figsize=(10,5))
sns.lineplot(x='year_month', y='monthly_revenue', data=bf , color = 'red')
plt.xticks(rotation=45)
plt.xlabel('Year Month')
plt.ylabel('Monthly Revenue')
plt.title('Monthly Revenue Over Time')
plt.show()
```

#Insights:
Revenue remains stable between 2.5M3.0M per month from JulyDecember 2025, with a clear peak in December (2.95M), indicating seasonal strength. The sharp drop in January 2026 (0.75M) reflects a partial month, not a decline in demand.

```python
sf = pd.read_csv('AVG.csv')
```

```python
sf
```

```python

import seaborn as sns

plt.figure(figsize=(10,5))
sns.barplot(x='Branch', y='revenue_per_invoice', data=sf, palette='dark', hue='Branch')
plt.xlabel('Branch')
plt.ylabel('Revenue Per Invoice')
plt.title('Revenue Per Invoice By Branch')
plt.show()
```

#Insights:
Muttathara generates the highest revenue per invoice (1,089) despite lower invoice volume, indicating stronger value per transaction. Palayam leads in scale (10.3K invoices) but with a lower average value (896), reflecting a high-volume model, while Counter Sales show unusually high averages due to low volume and should be excluded from performance benchmarking.

```python
sns.barplot(data = vf , x = 'Item_Category',y = 'revenue',hue = 'Item_Category',palette = 'bright')

plt.xlabel('Item Category')
plt.ylabel('Revenue')
plt.title('Revenue By Item Category')
plt.show()
```

```python
plt.pie(x=vf['revenue_pct'], labels=vf['Item_Category'], autopct='%1.1f%%', colors=sns.color_palette('bright'))
plt.title('Revenue By Item Category')
plt.show()
```

#Insights:
Spare parts generate 74.9% of total revenue despite a similar number of invoices as services, confirming they are the primary revenue engine. Services contribute 25.1% of revenue, indicating strong customer engagement but clear scope to increase value through pricing optimization or bundling with parts.
